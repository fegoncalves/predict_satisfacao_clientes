```python
#Importando os pacotes
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#Carregando os dados
dados = pd.read_csv('dadoshistoricos.csv')
```

```python
#Avaliando o tipo de objeto
type(dados)
```

```python
#Removendo a columa ID
dados = dados.drop('ID', axis=1)
dados.shape
```

```python
#exibindo os dados
dados.head(10)
```

```python
#Separando a variavel TARGET
TARGET = dados[['TARGET']]
```

```python
#Criando uma lista com as colunas do objeto, deixando a TARGET de fora (j est acima separada)
colunas = dados.columns[0:369].tolist()
```

```python
#Visualizando a lista com os nomes das colunas
colunas
```

```python
#Criando um array dos dados
array = dados.values
```

```python
#Visualizando o tipo do objeto
type(array)
```

```python
#Separando o array em componentes do input (x) da variavel TARGET
X = array[:,0:369]

#Gerando a nova ecala (normalizando os dados)
scaler = MinMaxScaler(feature_range = (0, 1))
rescaledX = scaler.fit_transform(X)
```

```python
#Transformando o Array em Dataframe
dadosNormalizados = pd.DataFrame(rescaledX, columns = colunas)
```

```python
#Visualizando o formato do dataframe
dadosNormalizados.shape
```

```python
#Reduzindo a dimensionalidade
pca = PCA(n_components = 30)
dados = pca.fit_transform(dadosNormalizados)
```

```python
#Convertendo o objeto PCA para um dataframe
dados = pd.DataFrame(dados)
```

```python
dados.head()
```

```python
#Concatetando a variavel TARGET ao DataFrame
dados = pd.concat([dados, TARGET], axis=1)
```

```python
#Visualizando os dados
dados
```

```python
#Dados de treino e de teste
dados_treino = dados.sample(frac=0.60, random_state=0)
dados_teste = dados.drop(dados_treino.index)
```

```python
#Gerando as variaveis preditoras e variavel TARGET
x_treino = dados_treino.iloc[:,0:30]
y_treino = dados_treino.iloc[:,30:31]

x_teste = dados_teste.iloc[:,0:30]
y_teste = dados_teste.iloc[:,30:31]
```

```python
#Contagem dos dados
dados_treino.count()
```

```python
#Contagem dos dados
dados_teste.count()
```

```python
#Classificador Random Forest
classificador = RandomForestClassifier(max_depth=5, random_state=0)
classificador.fit(x_treino, y_treino)

#Executando as previses
previsoes = classificador.predict(x_teste)
```

```python
#Avaliando o Confusion matrix e a acuracia
#Resultado  que a cada 100x, o modelo vai acertar 96%
print('Acuracia: %s' % accuracy_score(y_teste, previsoes))
```
